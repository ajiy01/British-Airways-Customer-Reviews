```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('BA_reviews_updated')
df
```

```python
## Removing Stopwords
import nltk 
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')

df['reviews_wo_stopwords'] = df['clean_reviews'].apply(lambda x: ' '.join([w for w in x.split() if w.lower() not in stopwords]))
df
```

```python
import re
df['reviews_wo_stopwords'] = df['reviews_wo_stopwords'].map(lambda x: re.sub('[,\.!?]', '', x))
df
```

```python
## Lemmatization
wn = nltk.WordNetLemmatizer()
nltk.download('wordnet')
nltk.download('omw-1.4')

df['reviews_lemmatized'] = df['reviews_wo_stopwords'].apply(lambda x: ' '.join([wn.lemmatize(w) for w in x.split() if w not in stopwords]))
df
```

```python
import gensim 
from gensim.utils import simple_preprocess
import nltk

def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))

data = df.reviews_lemmatized.values.tolist()
data_words = list(sent_to_words(data))
print(data_words[:1][0][:30])
```

```python
import gensim.corpora as corpora
id2word = corpora.Dictionary(data_words)

texts = data_words

corpus = [id2word.doc2bow(text) for text in texts]
print(corpus[:1][0][:30])
```

```python
from pprint import pprint
num_topics = 10

lda_model = gensim.models.LdaMulticore(corpus=corpus, id2word=id2word, num_topics=num_topics)

pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
import pyLDAvis
import pyLDAvis.gensim_models
pyLDAvis.enable_notebook()
panel = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word)
panel
```
